# Preprocessing

```python
sns.heatmap(final.isnull())
```

```python
print(final[['ROW_EFFECTIVE_TIMESTAMP',
       'ROW_EXPIRATION_TIMESTAMP', 'STORE_NAME', 'STORE_TYPE_CODE',
       'STORE_TYPE_NAME', 'CLOSE_DATE', 'CITY_NAME', 'STATE_CODE',
       'STATE_NAME', 'COUNTRY_CODE', 'TIME_ZONE_NAME', 'DMA_ID', 'DMA_NAME',
       'STORE_STATUS', 'REGION_ID', 'REGION_NAME', 'STORE_LIVE_DATE',
       'STORE_FUNCTIONAL_STATUS', 'GLOBAL_REGION_NAME',
       'GLOBAL_REGION_STORE_TYPE_NAME', 'AREA_REGION']])
```

```python
test = final.drop(['ROW_EFFECTIVE_TIMESTAMP','ROW_EXPIRATION_TIMESTAMP', 'STORE_NAME', 'STORE_TYPE_CODE',
       'STORE_TYPE_NAME','CITY_NAME', 'STATE_CODE','DMA_NAME','STORE_STATUS','REGION_NAME','TRANSACTION_FINAL_COUNT'
        'STATE_NAME', 'COUNTRY_CODE', 'TIME_ZONE_NAME','AREA_REGION','STORE_FUNCTIONAL_STATUS', 'GLOBAL_REGION_NAME','BUSINESS_DATE_FISCAL_DAY_OF_YEAR','EDW_CREATE_TIMESTAMP','EDW_MODIFY_TIMESTAMP'],axis=1)
```

```python
format='%m/%d/%Y'
df=test.drop(['OPEN_DATE','BUSINESS_DATE','CLOSE_DATE','STORE_LIVE_DATE','DISPATCH_GO_LIVE_DATE','DOOR_DASH_MARKUP_GO_LIVE_DATE','UBER_EATS_MARKUP_DATE'],axis=1)
test = pd.concat([df,test[[c for c in test.columns if c.split('_')[-1] == 'DATE']].map(lambda x: datetime.strptime(x,format))],axis=1)
```

```python
print(test['STORE_CONCEPT_NAME'].unique())
```

```python
test = pd.concat([test,pd.get_dummies(test['GLOBAL_REGION_STORE_TYPE_NAME'],drop_first=True,dtype=float)],axis=1)
test = pd.concat([test,pd.get_dummies(test['STORE_CONCEPT_NAME'],drop_first=True,dtype=float)],axis=1)
test.drop(['GLOBAL_REGION_STORE_TYPE_NAME','STORE_CONCEPT_NAME'],axis=1,inplace=True)
test
```

```python
for c in test.columns: 
    if c.split('_')[-1] == 'DATE':
        test[c+'_year'] = test[c].apply(lambda x: x.year)
        test[c+'_month'] = test[c].apply(lambda x: x.month)
        test[c+'_day'] = test[c].apply(lambda x: x.day)
        test.drop(c,inplace=True,axis=1)
test
```

# Training

```python
for c in test.columns:
    pd.to_numeric(test[c])
```

```python
test_train = pd.DataFrame(index=test.index,columns=test.columns)
test_test = pd.DataFrame(index=test.index,columns=test.columns)
for i in test['STORE_KEY'].unique():
    tempdf = test[test['STORE_KEY'] == i]
    test_test = pd.concat([test_test,tempdf[(tempdf['BUSINESS_DATE_year'] == 2023) & (tempdf['BUSINESS_DATE_month'] > 9)]],axis=0)
    test_train = pd.concat([test_train,tempdf[(tempdf['BUSINESS_DATE_year'] < 2023) | ((tempdf['BUSINESS_DATE_year'] == 2023) & (tempdf['BUSINESS_DATE_month'] <= 9))]],axis=0)
```

```python
test_train.dropna(inplace=True)
```

```python
test_test.dropna(inplace=True)
```

```python
test_train.columns
```

```python
X_train = test_train.drop(['NET_SALES_FINAL_USD_AMOUNT'],axis=1)
y_train = test_train['NET_SALES_FINAL_USD_AMOUNT']
X_test = test_test.drop(['NET_SALES_FINAL_USD_AMOUNT'],axis=1)
y_test = test_test['NET_SALES_FINAL_USD_AMOUNT']
```

```python
for c in X_train.columns:
    X_train[c]=pd.to_numeric(X_train[c])

y_train = pd.to_numeric(y_train)

for c in X_test.columns:
    X_test[c] = pd.to_numeric(X_test[c])

y_test = pd.to_numeric(y_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit_transform(X_train)
```

```python
scaler.transform(X_test)
```

```python
import xgboost as xgb
```

```python
reg = xgb.XGBRegressor(n_estimators=1000, early_stopping_rounds=50)
reg.fit(X_train,y_train,eval_set=[(X_train,y_train),(X_test,y_test)],verbose=True)
```

```python
test_test['prediction'] = reg.predict(X_test)
```
